```python
# read in a text file as a list of lines
def read_file_as_lines(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()
    return [line.strip() for line in lines]
```

```python
input = read_file_as_lines('input.txt')
```

```python
import numpy as np
# PART 1

voltages = []
for line in input: 
    line_split = [int(x) for x in line]

    # largest value except last
    first_digit_idx = np.argmax(line_split[0:-1])
    second_digit_idx = np.argmax(line_split[first_digit_idx+1:])
    
    # print(f"first idx: {first_digit_idx}, second idx: {second_digit_idx + first_digit_idx + 1}")
    voltages.append(int(str(line_split[first_digit_idx]) + str(line_split[second_digit_idx + first_digit_idx + 1])))
    
```

```python
sum(voltages)
```

```python
# PART 2

voltages = []
for line in input:
    line_split = [int(x) for x in line]
    start_idx = 0
    max_tracker = []
    for count in range(0, 12):
        finish_idx = len(line_split) - (12 - count) + 1
        window = line_split[start_idx:finish_idx]
        max_idx = np.argmax(window)
        start_idx += max_idx + 1
        max_tracker.append(window[max_idx])

    voltages.append(int(''.join([str(x) for x in max_tracker])))
sum(voltages)
```
