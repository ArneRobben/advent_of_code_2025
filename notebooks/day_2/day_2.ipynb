```python
# read in a text file as a list of lines
def read_file_as_lines(file_path):
    with open(file_path, 'r') as file:
        lines = file.read().strip()
    return lines.split(",")
```

```python
input = read_file_as_lines('input.txt')
```

```python
# part 1

def invalid(id: int) -> int:
    id = str(id)
    n = len(id)
    if id[0:n//2] == id[n//2:]:
        return int(id)
    else:
        return 0
```

```python
# part 2

def invalid(id:int)->int:
    id = str(id)

    for idx in range(0, len(id)-1):
        check = id.split(id[0:idx+1])
        # print(check)
        if sum([len(part) for part in check]) == 0:
            return int(id)

    else:
        return 0
```

```python
invalid_ids = 0
for combination in input:
    left, right = combination.split("-")

    for number in range(int(left), int(right)+1):
        # if invalid(number) != 0:
            # print(f"Invalid ID found: {number}")
        invalid_ids = invalid_ids + invalid(number)
```
