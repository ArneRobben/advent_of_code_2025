```python
# read in a text file as a list of lines
def read_file_as_lines(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()
    return [line.strip() for line in lines]
```

```python
input = read_file_as_lines('input.txt')
```

```python
ranges = input[0: input.index('')]
values = input[input.index('') + 1:]
```

```python
from tqdm import tqdm 
# PART 1

fresh_list = []

for range_line in tqdm(ranges):
    start, end = range_line.split('-')
    start = int(start)
    end = int(end)
    fresh_list.append(range(start, end + 1))
```

```python
fresh_count = 0
for value in values:
    num = int(value)
    for fresh_range in fresh_list:
        if num in fresh_range:
            fresh_count += 1
            break
```

```python
# Part 2

count = 0
already_parsed_ranges = []
for fresh_range in tqdm(fresh_list):
    for parsed_range in already_parsed_ranges:
        # there are four cases of overlap
        original_fresh_range = fresh_range
        # 1. fresh_range overlaps parsed_range on the right
        if fresh_range.start <= parsed_range.stop and fresh_range.stop >= parsed_range.stop and fresh_range.start >= parsed_range.start:
            # trim fresh_range
            fresh_range = range(parsed_range.stop, fresh_range.stop)
        # 2. fresh_range overlaps parsed_range on the left
        elif fresh_range.start <= parsed_range.start and fresh_range.stop <= parsed_range.stop and fresh_range.stop >= parsed_range.start:
            # trim fresh_range
            fresh_range = range(fresh_range.start, parsed_range.start)
        # 3. fresh_range is completely inside parsed_range
        elif fresh_range.start >= parsed_range.start and fresh_range.stop <= parsed_range.stop:
            # fresh range is completely covered, skip
            fresh_range = range(0,0)
        # 4. parsed_range is completely inside fresh_range
        elif fresh_range.start <= parsed_range.start and fresh_range.stop >= parsed_range.stop:
            # drop parsed_range from already_parsed_ranges
            already_parsed_ranges.remove(parsed_range)
        
        # print(f"parsed_range: {parsed_range}, original_fresh_range: {original_fresh_range}, new_fresh_range: {fresh_range}")


    # now add test_ranges to already_parsed_ranges
    if fresh_range.stop > fresh_range.start:
        already_parsed_ranges.append(fresh_range)

```

```python
# number of unique fresh ranges

fresh_sum = 0
for r in already_parsed_ranges:
    fresh_sum += (r.stop - r.start)
    # print(fresh_sum)
```
