```python
# read in a text file as a list of lines
def read_file_as_lines(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()
    return lines
```

```python
input = read_file_as_lines('sample.txt')
```

```python
input = [row.split(' ') for row in input]
input[:-1] = [[int(el) for el in row if el!=''] for row in input[:-1] ]
input[-1] = [el for el in input[-1] if el!='']
```

```python
import numpy as np

operator_results = []
for col in range(0, len(input[0])):
    col_values = [input[row][col] for row in range(0, len(input))]
    if col_values[-1] == '*':
        result = np.prod(col_values[:-1])
    else:
        result = np.sum(col_values[:-1])
    operator_results.append(result)
```

```python
input_processed = []
for idx in range(0, len(input_raw[0])):
    new_line = [input_raw[row][idx] for row in range(0, len(input_raw))]
    input_processed.append(new_line)
```
