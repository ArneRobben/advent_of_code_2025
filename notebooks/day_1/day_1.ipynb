```python
# read in a text file as a list of lines
def read_file_as_lines(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()
    return [line.strip() for line in lines]
```

```python
# PART 1
```

```python
input = read_file_as_lines('sample.txt')

dial = [50]

for rotation in input:
    direction = rotation[0]
    degrees = int(rotation[1:])

    if direction == 'L':
        dial.append((dial[-1] - degrees) % 100)
    elif direction == 'R':
        dial.append((dial[-1] + degrees) % 100)
```

```python
dial
```

```python
# count the number of zeros in dial

sum(1*[(d == 0) for d in dial])
```

```python
input = read_file_as_lines('sample.txt')

dial = [50]
passed_zero = 0

for rotation in input:
    direction = rotation[0]
    degrees = int(rotation[1:])

    if direction == 'L':
        new_dial = (dial[-1] - degrees)
    elif direction == 'R':
        new_dial = (dial[-1] + degrees)

    # count how many times we passed zero
    passed_zero += abs(new_dial // 100)
    # count if we landed on zero
    if new_dial % 100 == 0 and degrees != 0:
        passed_zero += 1
    else:   
        passed_zero += abs(new_dial // 100)
    dial.append(new_dial % 100)

    
```

```python
sum(1*[(d == 0) for d in dial])
```
