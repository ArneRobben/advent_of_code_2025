```python
# read in a text file as a list of lines
def read_file_as_lines(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()
    return [line.strip() for line in lines]
```

```python
# PART 1
```

```python
input = read_file_as_lines('sample.txt')

dial = [50]

for rotation in input:
    direction = rotation[0]
    degrees = int(rotation[1:])

    if direction == 'L':
        dial.append((dial[-1] - degrees) % 100)
    elif direction == 'R':
        dial.append((dial[-1] + degrees) % 100)
```

```python
dial
```

```python
# count the number of zeros in dial

sum(1*[(d == 0) for d in dial])
```

```python
from tqdm import tqdm
input = read_file_as_lines('input.txt')

dial = [50]
passed_zero = 0

for rotation in tqdm(input):
    direction = rotation[0]
    degrees = int(rotation[1:])
    current_dial = dial[-1]

    for _ in range(degrees):
        if direction == 'L':
            current_dial = (current_dial - 1) % 100
        elif direction == 'R':
            current_dial = (current_dial + 1) % 100

        if current_dial == 0:
            passed_zero += 1
    dial.append(current_dial)
    
```
